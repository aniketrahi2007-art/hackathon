```python
import requests, re, json
from datetime import datetime, timezone
import uuid

SANDBOX_URL = "https://script.google.com/macros/s/AKfycbyIGT-uwrzizHEn94jCl3Q8iTDStZOtM7tLV2cH7iqpauZtqy0OGDslcYMv_HlLd46J/exec"

def tokenize(text):
    return set(re.findall(r"[a-zA-Z0-9\+#\.]+", (text or "").lower()))

# 1) Fetch jobs
jobs = requests.get(SANDBOX_URL, params={"path":"api/jobs"}, timeout=30).json()["jobs"]
print("Jobs fetched:", len(jobs))

# 2) Fetch student profile row(s)
profiles = requests.get(SANDBOX_URL, params={"path":"api/table", "name":"STUDENT_PROFILE"}, timeout=30).json()["rows"]
profile = [p for p in profiles if str(p.get("student_id","")) == "student_demo_001"][0]
facts_profile = json.loads(profile["facts_profile_json"])
student_skills = set(s.lower() for s in facts_profile.get("skills", []))
print("Student skills:", student_skills)

# 3) Fetch bullets
bullets = requests.get(SANDBOX_URL, params={"path":"api/table", "name":"BULLET_BANK"}, timeout=30).json()["rows"]
bullets = [b for b in bullets if str(b.get("student_id","")) == "student_demo_001"]
print("Bullets fetched:", len(bullets))

# 4) Score jobs with explanations
ranked = []
for j in jobs:
    text = f"{j.get('title','')} {j.get('requirements_text','')} {j.get('description_text','')}"
    tokens = tokenize(text)

    overlap = sorted([s for s in student_skills if s in tokens])
    skill_score = len(overlap) / max(1, len(student_skills))

    remote_ok = str(j.get("remote","")).strip().lower() in ["true","yes","1"]
    constraint_score = 1.0 if remote_ok else 0.0  # we enforce remote for demo

    total = round(0.7*skill_score + 0.3*constraint_score, 3)

    explanation = {
        "skill_overlap": overlap,
        "skill_score": round(skill_score, 3),
        "remote_ok": remote_ok,
        "constraint_score": round(constraint_score, 3),
        "total_score": total
    }
    ranked.append((total, j, explanation))

ranked.sort(key=lambda x: x[0], reverse=True)
print("Top 5 scores:", [r[0] for r in ranked[:5]])
```

```python
import requests, json, re
from datetime import datetime, timezone

SANDBOX_URL = "https://script.google.com/macros/s/AKfycbyIGT-uwrzizHEn94jCl3Q8iTDStZOtM7tLV2cH7iqpauZtqy0OGDslcYMv_HlLd46J/exec"
STUDENT_ID = "student_demo_001"

def tokenize(text):
    return set(re.findall(r"[a-zA-Z0-9\+#\.]+", (text or "").lower()))

# Fetch jobs
jobs = requests.get(SANDBOX_URL, params={"path":"api/jobs"}, timeout=30).json()["jobs"]
jobs_by_id = {str(j["job_id"]): j for j in jobs}

# Fetch bullets
bullets_rows = requests.get(SANDBOX_URL, params={"path":"api/table", "name":"BULLET_BANK"}, timeout=30).json()["rows"]
bullets_rows = [b for b in bullets_rows if str(b.get("student_id","")) == STUDENT_ID]

# Fetch queue
queue_rows = requests.get(SANDBOX_URL, params={"path":"api/table", "name":"APPLY_QUEUE"}, timeout=30).json()["rows"]
queue_rows = [q for q in queue_rows if str(q.get("student_id","")) == STUDENT_ID]

# Build a simple evidence index: map token -> best bullet(s)
token_to_bullets = {}
for b in bullets_rows:
    bt = (b.get("bullet_text") or "").lower()
    bt_tokens = tokenize(bt)
    for t in bt_tokens:
        token_to_bullets.setdefault(t, []).append({
            "bullet_id": b.get("bullet_id"),
            "bullet_text": b.get("bullet_text"),
            "source_name": b.get("source_name"),
        })

def requirement_evidence_map(job):
    text = f"{job.get('requirements_text','')} {job.get('description_text','')}".lower()
    tokens = sorted(list(tokenize(text)))

    # We only map a limited set of meaningful “requirements”
    # (otherwise every random word becomes a requirement)
    candidate_reqs = [t for t in tokens if t in {"python","sql","excel","react","api","apis","machine","learning","statistics","power","bi","dashboard","dashboards","etl","testing","git"}]

    mapped = []
    for req in candidate_reqs:
        if req in token_to_bullets:
            # pick the first bullet as evidence (simple + explainable)
            ev = token_to_bullets[req][0]
            mapped.append({"requirement": req, "evidence_type": "bullet", "evidence": ev})
        else:
            mapped.append({"requirement": req, "evidence_type": "no_evidence", "evidence": None})

    return mapped

# Update each queue row with requirement->evidence JSON
now = datetime.now(timezone.utc).isoformat()
updated_queue = []

for q in queue_rows:
    job_id = str(q.get("job_id",""))
    job = jobs_by_id.get(job_id)
    if not job:
        q["requirements_evidence_json"] = "[]"
    else:
        mapping = requirement_evidence_map(job)
        q["requirements_evidence_json"] = json.dumps(mapping)
    q["created_at"] = q.get("created_at") or now
    updated_queue.append(q)

payload = {
    "sheet": "APPLY_QUEUE",
    "key_col": "student_id",
    "key_val": STUDENT_ID,
    "rows": updated_queue
}

resp = requests.post(
    SANDBOX_URL,
    params={"path":"api/replace_rows"},
    data=json.dumps(payload),
    headers={"Content-Type":"application/json"},
    timeout=30
)

print("STATUS:", resp.status_code)
print("RESPONSE:", resp.text)

# Print one example mapping for a top job (first row)
if updated_queue:
    print("\nExample job_id:", updated_queue[0]["job_id"])
    example = json.loads(updated_queue[0]["requirements_evidence_json"])
    print("Example mapping (first 6):", example[:6])
```

```python
import requests, json, time, hashlib
from datetime import datetime, timezone

SANDBOX_URL = "https://script.google.com/macros/s/AKfycbyIGT-uwrzizHEn94jCl3Q8iTDStZOtM7tLV2cH7iqpauZtqy0OGDslcYMv_HlLd46J/exec"
STUDENT_ID = "student_demo_001"

def get_table(name):
    return requests.get(SANDBOX_URL, params={"path":"api/table","name":name}, timeout=30).json()["rows"]

def sha16(obj):
    return hashlib.sha256(json.dumps(obj, sort_keys=True).encode("utf-8")).hexdigest()[:16]

def read_policy():
    rows = get_table("APPLY_POLICY")
    policy = {}
    for r in rows:
        k = str(r.get("policy_key","")).strip()
        if not k:
            continue
        policy[k] = r.get("policy_value")
    # normalize
    def as_bool(v): return str(v).strip().upper() == "TRUE"
    def as_int(v, d):
        try: return int(str(v).strip())
        except: return d
    def as_float(v, d):
        try: return float(str(v).strip())
        except: return d
    def as_list_json(v):
        try: return json.loads(v) if v else []
        except: return []

    return {
        "kill_switch": as_bool(policy.get("kill_switch","FALSE")),
        "max_applications_per_run": as_int(policy.get("max_applications_per_run","10"), 10),
        "min_match_score": as_float(policy.get("min_match_score","0.55"), 0.55),
        "blocked_companies": as_list_json(policy.get("blocked_companies_json","[]")),
        "blocked_role_keywords": [s.lower() for s in as_list_json(policy.get("blocked_role_keywords_json","[]"))],
        "cooldown_seconds": as_int(policy.get("cooldown_seconds","1"), 1),
    }

def post_apply(payload, max_retries=4):
    attempt = 0
    while True:
        attempt += 1
        r = requests.post(
            SANDBOX_URL,
            params={"path":"api/apply"},
            data=json.dumps(payload),
            headers={"Content-Type":"application/json"},
            timeout=30
        )
        data = r.json()

        if data.get("ok"):
            return {"ok": True, "receipt": data["receipt"], "attempts": attempt}

        # retry conditions (simulated by server)
        err = data.get("error","")
        if attempt >= max_retries or err == "KILL_SWITCH_ON":
            return {"ok": False, "error": err, "attempts": attempt}

        # backoff
        wait_s = int(data.get("retry_after_sec", 2)) + attempt
        time.sleep(wait_s)

def log_failure(job, submitted_at, retry_count, reason, package):
    # optional endpoint; if you didn't add it, this will fail safely
    payload = {
        "job_id": job["job_id"],
        "company": job.get("company",""),
        "title": job.get("title",""),
        "student_id": STUDENT_ID,
        "status": "failed",
        "submitted_at": submitted_at,
        "retry_count": retry_count,
        "failure_reason": reason,
        "payload_hash": sha16(package),
        "package_json": json.dumps(package)
    }
    try:
        r = requests.post(
            SANDBOX_URL,
            params={"path":"api/log_application_attempt"},
            data=json.dumps(payload),
            headers={"Content-Type":"application/json"},
            timeout=30
        )
        return r.json()
    except Exception as e:
        return {"ok": False, "error": str(e)}

# Load policy
policy = read_policy()
print("APPLY POLICY:", policy)

if policy["kill_switch"]:
    print("Kill switch is ON. Stopping.")
    raise SystemExit

# Load queue + jobs
queue = get_table("APPLY_QUEUE")
queue = [q for q in queue if str(q.get("student_id","")) == STUDENT_ID]
queue.sort(key=lambda q: float(q.get("match_score") or 0), reverse=True)

jobs = requests.get(SANDBOX_URL, params={"path":"api/jobs"}, timeout=30).json()["jobs"]
jobs_by_id = {str(j["job_id"]): j for j in jobs}

# Already applied job_ids
apps = get_table("APPLICATIONS")
already = set(str(a.get("job_id","")) for a in apps if str(a.get("student_id","")) == STUDENT_ID and str(a.get("status","")) == "submitted")

applied = 0
for q in queue:
    if applied >= policy["max_applications_per_run"]:
        break

    score = float(q.get("match_score") or 0)
    if score < policy["min_match_score"]:
        continue

    job_id = str(q.get("job_id",""))
    job = jobs_by_id.get(job_id)
    if not job:
        continue

    if job_id in already:
        continue

    company = str(job.get("company",""))
    title = str(job.get("title","")).lower()

    if company in policy["blocked_companies"]:
        continue
    if any(k in title for k in policy["blocked_role_keywords"]):
        continue

    # Build package (grounded)
    requirements_map = q.get("requirements_evidence_json") or "[]"
    package = {
        "student_id": STUDENT_ID,
        "job_id": job_id,
        "company": company,
        "title": job.get("title",""),
        "tailored_summary": f"Applying for {job.get('title','')} at {company}. This application only uses facts from the stored artifact pack.",
        "requirement_evidence_map": json.loads(requirements_map),
        "notes": "No credentials or numbers are invented. Requirements without evidence are marked no_evidence."
    }

    submitted_at = datetime.now(timezone.utc).isoformat()

    payload = {
        "job_id": job_id,
        "company": company,
        "title": job.get("title",""),
        "student_id": STUDENT_ID,
        "submitted_at": submitted_at,
        "retry_count": 0,
        "package": package
    }

    result = post_apply(payload, max_retries=4)

    if result["ok"]:
        applied += 1
        receipt = result["receipt"]["receipt_id"]
        print(f"✅ SUBMITTED {applied}/ {policy['max_applications_per_run']}  job={job_id}  score={score:.3f}  receipt={receipt}")
    else:
        print(f"❌ FAILED job={job_id} after {result['attempts']} attempts  reason={result['error']}")
        # log failure if endpoint exists
        log_res = log_failure(job, submitted_at, result["attempts"], result["error"], package)
        print("   failure logged:", log_res)

    time.sleep(policy["cooldown_seconds"])

print("\nDONE. Applications submitted:", applied)
```
